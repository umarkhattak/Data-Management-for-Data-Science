```python
'''
3. [5 pts] The values in the Android version column should be floats. 
Strip the trailing non-numeric characters from all values (ie. the 
words " and up"), so the result is a number. If there are multiple 
decimal places (eg. "x.y.z"), keep only the first two parts (eg "x.y"). 
For example, the value "4.1 and up" should be changed to "4.1". The value 
"4.5.6 and up" should be changed to "4.5". The value "5.6.7" should be 
changed to "5.6".
'''
dfpd['Android Ver'] = dfpd['Android Ver'].str.extract('(\d+\.?\d?)', expand = False)
dfpd['Android Ver'] = (dfpd['Android Ver']).apply(float)

dfpd
```

```python
'''
4. [5 pts] The "Installs" column must have integer values. 
For values that have commas, remove the commas. 
For values that have a '+' at the end, remove the 
'+'. Keep only those rows that have an integer value 
after these edits.
'''
dfpd['Installs']=(dfpd['Installs']).apply(str)
dfpd['Installs'] = dfpd['Installs'].str.replace(",","")
dfpd['Installs'] = dfpd['Installs'].apply(lambda l: l.rstrip('+'))
dfpd = dfpd[dfpd['Installs'].apply(lambda l: l.isnumeric())] 

dfpd
```

```python
'''
5. [5 pts] For missing rating values, if the number of reviews is 
less than 100 and installations is less than 50000, remove the row. 
Else, fill the missing value with the average value (rounded to 2 
decimal places) for the Category of that row.
'''
cat = dfpd.groupby(['Category'], as_index = False).mean()
def questionThreePreprocessingPartFive(dfpd_category, dfpd_rating):
    if pd.isnull(dfpd_rating) == True:
        sCat = cat.loc[cat['Category'] == dfpd_category]  
        dfpd_rating = float(round(sCat['Rating'], 2))
        return dfpd_rating
    else:
        return dfpd_rating
    
dfpd['Installs']=(dfpd['Installs']).apply(float)
dfpd['Reviews'] = (dfpd["Reviews"]).apply(float)

newDf = dfpd[((dfpd['Installs']) < 50000) & ((dfpd['Reviews']) < 100) & (dfpd['Rating'].isnull() == True)].index
dfpd.drop(newDf, inplace = True)

dfpd["Rating"] = dfpd.apply(lambda l: questionThreePreprocessingPartFive(l["Category"], l["Rating"]), axis = 1)
dfpd
```
