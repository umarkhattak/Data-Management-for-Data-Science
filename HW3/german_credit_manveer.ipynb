## Problem 2 Preprocessing

###  1.[8 pts] Drop the 3 columns that contribute the least to the dataset. These would be the columns with the highest number of non-zero 'none' values. Break ties by going left to right in columns. (Your code should be generalizable to drop n columns, but for the rest of the analysis, you can call your code for n=3.)

```python
germ = pd.read_csv("GermanCredit.csv")
col_dic = defaultdict(list)
#print(list(germ.columns))
def drop_n_columns(germ, n = 3):    
    for col in list(germ.columns):
         num_of_nones = germ[germ[col] == 'none'].shape[0]      
         col_dic[num_of_nones].append(col)
    col_list  = list(col_dic.items())
    col_list.sort(key = lambda x: x[0], reverse = True)
    #print(col_list)
    #print("\n")
    
    fin_list_1 = []
    for item in list(dict(col_list).values()):
        fin_list_1.extend(item)
    fin_list_2 = fin_list_1[0:n]
    
    for col in fin_list_2:
        del germ[col]
        print(col)
    
    
  
```

### 2 [4 pts] Certain values in some of the columns contain unnecessary apostrophes (‘). Remove the apostrophes.

```python
#[a for a in germ['checking_status'] if a.find("'") != -1]

for row_ind in range(germ.shape[0]):
    
    for col_ind in range(germ.shape[1]):
        
        if str(germ.iloc[row_ind,col_ind]).find("'") != -1:
                         
            germ.iloc[row_ind,col_ind] =  str(germ.iloc[row_ind,col_ind]).replace("'", '')
    
```

### 3 [5 pts] The checking_status column has values in 4 categories: 'no checking', '<0', '0<=X<200', and '>=200'. Change these to 'No Checking', 'Low', 'Medium', and 'High' respectively.

```python


for i in range(germ.shape[0]):
    
     val = germ.loc[i,"checking_status"]
     
     if(val == 'no checking'): 
        germ.loc[i,"checking_status"] = 'No Checking'
                
     elif val == '<0': 
        germ.loc[i,"checking_status"] = 'Low'
    
     elif val == '0<=X<200': 
        germ.loc[i,"checking_status"] = 'Medium'
            
        
     elif val == '>=200':
        germ.loc[i,"checking_status"] = 'High'
        
    
    
#germ.head(10)
#df["checking_status"] = "No Checking"
```

```python
for i in range(germ.shape[0]):
    
     val = germ.loc[i,"savings_status"]
     
     if(val == 'no known savings'): 
        germ.loc[i,"savings_status"] = 'No Savings'
                
     elif val == '<100': 
        germ.loc[i,"savings_status"] = 'Low'
    
     elif val == '100<=X<500': 
        germ.loc[i,"savings_status"] = 'Medium'
            
        
     elif val == '500<=X<1000' or val == '>=1000':
        germ.loc[i,"savings_status"] = 'High'
```

### 4 [4 pts] For the purpose 'education', what is the most common checking_status and savings_status? Your code should print:

### 3 [6 pts] For people with a "High" savings_status and age above 40, use subplots to plot the following pie charts: personal status, credit history, and job
